## 1. Introducción

BYMAX es un sistema web de orientación diagnóstica preliminar inspirado en la idea de un asistente conversacional de apoyo en salud. En su estado actual, el proyecto funciona como un asistente que interactúa con el usuario mediante preguntas sobre su perfil y sus síntomas, y a partir de esta información realiza una inferencia basada en un conjunto de datos clínicos estructurados.

Actualmente, **Baymax** actúa como un agente conversacional simple con lógica basada en reglas y filtrado estadístico sobre el dataset. Su función principal es:

- recopilar información del usuario,
- procesar edad, género y síntomas,
- consultar patrones dentro del dataset,
- estimar enfermedades probables,
- presentar un resultado orientativo con apoyo visual.

En esta fase del proyecto, Baymax **es un asistente médico orientativo**, no un sistema clínico real ni un modelo médico certificado. Más adelante, el proyecto podrá evolucionar hacia una arquitectura más robusta basada en agentes especializados y modelos computacionales más avanzados.

```python
# Ajusta esta ruta según subas el CSV a Colab
ruta_csv = 'Disease_symptom_and_patient_profile_dataset.csv'
df = pd.read_csv(ruta_csv)
df.head()
```

## 9. Exploración inicial de la data

En esta sección revisamos la estructura general del dataset: número de filas, columnas, tipos de datos y valores faltantes. Esto es importante para entender el estado original de la información antes del proceso de limpieza.


```python
print('Dimensiones del dataset:', df.shape)
print('\nColumnas:')
print(df.columns.tolist())
print('\nTipos de datos:')
print(df.dtypes)
print('\nValores nulos por columna:')
print(df.isnull().sum())
```

## 10. Limpieza y transformación de datos

En el proyecto actual, la limpieza de datos incluye:

- renombrar columnas del inglés al español,
- transformar valores `Yes/No` a `Sí/No`,
- eliminar filas con valores nulos,
- reiniciar índices.

Esto mejora la interpretabilidad de la data y facilita la interacción del usuario con el sistema en español.


```python
COLUMNAS_ES = {
    'Disease': 'Enfermedad',
    'Fever': 'Fiebre',
    'Cough': 'Tos',
    'Fatigue': 'Fatiga',
    'Difficulty Breathing': 'Dificultad al respirar',
    'Age': 'Edad',
    'Gender': 'Género',
    'Blood Pressure': 'Presión arterial',
    'Cholesterol Level': 'Colesterol',
    'Outcome Variable': 'Resultado'
}

VALORES_ES = {'Yes': 'Sí', 'No': 'No'}
SINTOMAS = ['Fiebre', 'Tos', 'Fatiga', 'Dificultad al respirar']

df_limpio = df.rename(columns=COLUMNAS_ES).copy()
for col in SINTOMAS:
    if col in df_limpio.columns:
        df_limpio[col] = df_limpio[col].map(VALORES_ES)

df_limpio = df_limpio.dropna().reset_index(drop=True)
df_limpio.head()
```

## 11. Verificación posterior a la limpieza

Después del preprocesamiento, verificamos el estado final del dataset para confirmar que está listo para análisis y uso dentro del sistema.


```python
print('Dimensiones después de limpieza:', df_limpio.shape)
print('\nValores nulos restantes:')
print(df_limpio.isnull().sum())
print('\nCantidad de enfermedades únicas:', df_limpio['Enfermedad'].nunique())
```

## 12. Análisis exploratorio de la data

En esta fase buscamos comprender cómo está distribuida la información en el dataset, qué enfermedades aparecen con mayor frecuencia y qué síntomas tienen mayor presencia.


```python
resumen = {
    'total_registros': len(df_limpio),
    'enfermedades_unicas': df_limpio['Enfermedad'].nunique(),
    'con_fiebre': int((df_limpio['Fiebre'] == 'Sí').sum()),
    'con_tos': int((df_limpio['Tos'] == 'Sí').sum())
}
resumen
```

```python
frecuencia_sintomas = {s: int((df_limpio[s] == 'Sí').sum()) for s in SINTOMAS}
frecuencia_sintomas
```

```python
top_enfermedades = df_limpio['Enfermedad'].value_counts().head(10)
top_enfermedades
```

## 13. Graficación de la información

La representación gráfica permite interpretar con mayor facilidad la distribución de los síntomas y las enfermedades más frecuentes. En el proyecto original, estas visualizaciones también se integran a la interfaz web para apoyar el diagnóstico orientativo.


```python
plt.figure(figsize=(8,4))
plt.bar(frecuencia_sintomas.keys(), frecuencia_sintomas.values(), color='teal')
plt.title('Frecuencia de síntomas en el dataset')
plt.ylabel('Cantidad de registros')
plt.xticks(rotation=15)
plt.show()
```

```python
plt.figure(figsize=(10,5))
top_enfermedades.sort_values().plot(kind='barh', color='mediumseagreen')
plt.title('Top 10 enfermedades más frecuentes')
plt.xlabel('Número de casos')
plt.show()
```

```python
if 'Edad' in df_limpio.columns:
    plt.figure(figsize=(8,4))
    sns.histplot(df_limpio['Edad'], kde=True, color='royalblue')
    plt.title('Distribución de la edad en el dataset')
    plt.show()
```

## 14. Lógica actual del sistema

El funcionamiento actual de Baymax puede resumirse así:

1. Se crea una sesión por usuario.
2. Se recopilan edad y género.
3. Se preguntan síntomas uno por uno.
4. El sistema filtra el dataset con base en las respuestas.
5. Calcula las enfermedades con mayor coincidencia.
6. Si alcanza un umbral de confianza o se completan las preguntas, muestra un diagnóstico orientativo.

Esto significa que el sistema actual no usa todavía modelos avanzados de aprendizaje automático, sino una combinación de reglas, filtrado y conteo de frecuencia para tomar decisiones.
